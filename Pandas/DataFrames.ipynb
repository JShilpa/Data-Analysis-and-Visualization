```python
# Adding New column and assigning value as None.

ipl_frame['State'] = 'None'

ipl_frame
```

```python
# Assigning values through series

states = ['Maharashtra', 'Karnataka', 'West Bengal', 'Punjab', 'Chennai', 'Rajasthan']

values = pd.Series(states, index=[0,1,2,3,5,6])

ipl_frame['State'] = values

ipl_frame
```

```python
# To delete column State from the table

del ipl_frame['State']

ipl_frame
```
